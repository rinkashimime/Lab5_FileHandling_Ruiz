```python
# Step 1: import additional libraries
from datetime import datetime
import shutil
```

```python
# Step 2: Define working directory
backup_dir = Path.home() / "Documents" / "PythonFiles"
backup_dir.mkdir(exist_ok=True) 
```

```python
# Step 3: Define a function to write a file with backup
def write_with_backup(filename: str, content: str, mode: str = "overwrite"):
    file_path = backup_dir / filename
    # Create a backup if the file already exists
    if file_path.exists():
        timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
        backup_path = file_path.with_name(f"{file_path.stem}_backup_Ruiz_{timestamp}{file_path.suffix}")
        shutil.copy2(file_path, backup_path)
        print(f"Backup saved: {backup_path.name}")
    # Write the new content
    if mode == "append" and file_path.exists():
        with open(file_path, "a", encoding="utf-8") as file:
            file.write("\n" + content)
    else:
        with open(file_path, "w", encoding="utf-8") as file:
            file.write(content)
    print(f"File saved ({mode} mode): {file_path.name}")
```

```python
# Step 4: Define a function to read the file content
def read_file(filename: str):
    file_path = backup_dir / filename
    with open(file_path, "r", encoding="utf-8") as file:
        content = file.read()
        return content

# Step 4b: Define a function to list backups for a specific file
def list_backups(filename: str):
    """List all backup versions of a given filename"""
    stem = Path(filename).stem
    backups = sorted(backup_dir.glob(f"{stem}_backup_Ruiz_*"))
    if backups:
        print(f"Backups for {filename}:")
        for backup in backups:
            print(f"  - {backup.name}")
        return backups
    else:
        print(f"No backups found for {filename}")
        return []
```

```python
# Step 5: Run demonstration
print("=== File Operations Demo ===")

print("\n1. Creating new file:")
write_with_backup("demo.txt", "Initial Content")

print("\n2. Updating file (with backup):")
write_with_backup("demo.txt", "Updated Content")

print("\n3. Appending to file (with backup):")
write_with_backup("demo.txt", "Appended Line", mode="append")

print("\n4. Reading file:")
print("Current content:")
print(read_file("demo.txt"))

print("\n5. Listing all backups for demo.txt:")
list_backups("demo.txt")

print("\n6. Listing all backups in directory:")
for back in sorted(backup_dir.glob("*backup*")):
    print("-", back.name)
```

```python
# Step 3 Define the menu-driven file manager
def file_manager():
    file_name = input("Enter filename (e.g., notes.txt): ")
    file_path = backup_dir / file_name

    while True:
        print("\n--- MENU ---")
        print("1. Write to File")
        print("2. Append to File")
        print("3. Read File")
        print("4. Backup File")
        print("5. List Backups for This File")
        print("6. Exit")

        choice = input("Enter your choice (1-6): ")

        if choice == "1":
            content = input("Enter content to write:\n")
            with open(file_path, "w", encoding="utf-8") as file:
                file.write(content)
            print("File written successfully.")

        elif choice == "2":
            more = input("Enter content to append:\n")
            with open(file_path, "a", encoding="utf-8") as file:
                file.write("\n" + more)
            print("Content appended.")
        
        elif choice == "3":
            if file_path.exists():
                with open(file_path, "r", encoding="utf-8") as file:
                    print("\nFILE CONTENT:\n", file.read())
            else:
                print("File not found.")

        elif choice == "4":
            if file_path.exists():
                timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
                backup_path = file_path.with_name(f"{file_path.stem}_backup_Ruiz_{timestamp}{file_path.suffix}")
                shutil.copy2(file_path, backup_path)
                print(f"Backup created: {backup_path.name}")
            else:
                print("Cannot backup. File does not exist.")

        elif choice == "5":
            stem = Path(file_name).stem
            backups = sorted(backup_dir.glob(f"{stem}_backup_Ruiz_*"))
            if backups:
                print(f"\nBackups for {file_name}:")
                for backup in backups:
                    print(f"  - {backup.name}")
            else:
                print(f"No backups found for {file_name}")
            
        elif choice == "6":
            print("Exiting file manager.")
            break   

        else:
            print("Invalid choice. Please try again.")
```

```python
# Step 6: Summary Report - Directory Structure and Backup Registry

# Import required modules and define backup_dir
from pathlib import Path
backup_dir = Path.home() / "Documents" / "Ruiz_Activity_5"

print("\n" + "=" * 70)
print("FINAL REPORT: Ruiz_Activity_5 Directory Structure & Backups")
print("=" * 70)

# Display directory info
print(f"\n📁 Working Directory: {backup_dir}")
print(f"✓ Directory exists: {backup_dir.exists()}")

# Count files by type
all_files = list(backup_dir.glob("*"))
txt_files = [f for f in all_files if f.suffix == ".txt"]
md_files = [f for f in all_files if f.suffix == ".md"]
csv_files = [f for f in all_files if f.suffix == ".csv"]
backup_files = [f for f in all_files if "_backup_Ruiz_" in f.name]

print(f"\n📊 File Statistics:")
print(f"   Total files: {len(all_files)}")
print(f"   .txt files: {len(txt_files)}")
print(f"   .md files: {len(md_files)}")
print(f"   .csv files: {len(csv_files)}")
print(f"   Backup files: {len(backup_files)}")

# List current files (non-backup)
print(f"\n📄 Current Files (Active Data):")
current_files = sorted([f for f in all_files if "_backup_Ruiz_" not in f.name])
for file in current_files:
    size = file.stat().st_size
    print(f"   • {file.name} ({size} bytes)")

# List backups organized by original file
print(f"\n🔄 Backup Files by Source File:")
if backup_files:
    for orig_file in current_files:
        base_name = orig_file.stem
        matching_backups = sorted([f for f in backup_files if base_name in f.name])
        if matching_backups:
            print(f"   {orig_file.name}:")
            for backup in matching_backups:
                print(f"      └─ {backup.name}")
else:
    print("   No backup files found in this session")

print("\n" + "=" * 70)
```
